```python
import cv2
import os
import random
import numpy as np

from IPython.display import Image, display

from sklearn.decomposition import PCA

from skimage import exposure
```

```python
imgs_dir = "C:\Users\Erroll\Documents\eyespose\eyespose"
```

### Make a balanced dataset

Balanced image filenames stored in `all_imnames_balanced`

```python
all_imnames = os.listdir(imgs_dir)

# get filnames for lock and no-lock images
nolock_fns = [fn for fn in all_imnames if "nolock" in fn]
lock_fns = set(all_imnames) - set(nolock_fns)

# number of images in each class for balanced set
num_ims = min(len(lock_fns), len(nolock_fns))

# make a balanced dataset
all_imnames_balanced = random.sample(nolock_fns, num_ims) + random.sample(lock_fns, num_ims)
random.shuffle(all_imnames_balanced);

print "total number of images: %d (%d eye-lock)"%(len(all_imnames_balanced), num_ims)
```

### Load all images into a matrix

Rows of the matrix are the flattened images.
Show the mean eye image and some random eye images.

```python
# read in all images and equalize histograms
all_ims = [cv2.imread(os.path.join(imgs_dir, imname), cv2.CV_LOAD_IMAGE_GRAYSCALE) for imname in all_imnames_balanced]
all_imgs = [exposure.equalize_hist(img)*255.0 for img in all_ims]

all_ims_flat = [im.flatten() for im in all_ims]
matrix = np.vstack(all_ims_flat)
immean = matrix.mean(axis=0)

print "%d images, each has %d pixels"%matrix.shape

display(Image(data=cv2.imencode(".png", immean.reshape(24,40))[1].tostring()))
display(Image(data=cv2.imencode(".png", np.hstack(random.sample(all_ims,20)))[1].tostring()))
```

```python
pca = PCA()
pca.fit(all_ims_flat)

print pca
print pca.explained_variance_ratio_

print sum(pca.explained_variance_ratio_[:5])

for eigen_vec in pca.components_[:16]:
    print np.linalg.norm(eigen_vec)
```

```python
print pca.components_

princom_imgs = []

for eigen_vec in pca.components_[:16]:
    eigen_vec_img = eigen_vec.reshape(24,40).astype(np.float32)
    cv2.normalize(eigen_vec_img, eigen_vec_img, 0, 255, cv2.NORM_MINMAX)
    princom_imgs.append(eigen_vec_img)

Image(data=cv2.imencode(".png", np.hstack(princom_imgs))[1].tostring())
```

```python
idxs = random.sample(range(len(all_ims_flat)), 16)

orig_imgs, reconstructed_imgs = [], []

for idx in idxs:
    
    coeffs = [np.dot(all_ims_flat[idx] - immean, v) for v in pca.components_[:16]]

    reconstructed_img = immean.copy()
    for (coeff, vec) in zip(coeffs,pca.components_[:16]):
        reconstructed_img += coeff * vec
        
    orig_imgs.append(all_ims_flat[idx].reshape(24,40))
    reconstructed_imgs.append(reconstructed_img.reshape(24,40).reshape(24,40))

display(Image(data=cv2.imencode(".png", np.hstack(orig_imgs))[1].tostring()))
display(Image(data=cv2.imencode(".png", np.hstack(reconstructed_imgs))[1].tostring()))
```
