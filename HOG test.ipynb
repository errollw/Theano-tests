### Get random sets of filenames

```python
all_imnames = os.listdir(imgs_dir)
# all_imnames = [imname for imname in os.listdir(dataset_dir_single) if "_g" in imname]

nolock_fns = [fn for fn in all_imnames if "nolock" in fn]
lock_fns = set(all_imnames) - set(nolock_fns)
num_lock_imgs = len(lock_fns)
nolock_fns_balanced = random.sample(nolock_fns, num_lock_imgs)

all_imnames_balanced = list(set(nolock_fns_balanced).union(lock_fns))

random_nolock_fns = random.sample(nolock_fns_balanced, 16)
random_lock_fns = random.sample(lock_fns, 16)
```

### Visualize HOG for eye images

```python
def get_hog_imgs(list_of_fns):
    
    eye_imgs, hog_imgs = [], []
    
    for imname in list_of_fns:
        img = cv2.imread(os.path.join(imgs_dir, imname), cv2.CV_LOAD_IMAGE_GRAYSCALE)
        img = cv2.resize(img, (80,48))
        fd, hog_image = hog(img/255.0, orientations=9, pixels_per_cell=(8, 8), cells_per_block=(1, 1), visualise=True)
        hog_image_rescaled = exposure.rescale_intensity(hog_image, out_range=(0, 255))
        sep = np.ones((img.shape[0],2))*255
        eye_imgs.append(np.hstack([img,sep]))
        hog_imgs.append(np.hstack([hog_image_rescaled,sep]))
        
    return eye_imgs, hog_imgs

nolock_eye_imgs, nolock_hog_imgs = get_hog_imgs(random_nolock_fns)
lock_eye_imgs, lock_hog_imgs = get_hog_imgs(random_lock_fns)
    
cv2.imwrite(os.path.join(debug_dir,"nolock_eye_imgs.png"), np.hstack(nolock_eye_imgs))
display(Image(os.path.join(debug_dir,"nolock_eye_imgs.png")))
    
cv2.imwrite(os.path.join(debug_dir,"nolock_hog_imgs.png"), np.hstack(nolock_hog_imgs))
display(Image(os.path.join(debug_dir,"nolock_hog_imgs.png")))

cv2.imwrite(os.path.join(debug_dir,"lock_eye_imgs.png"), np.hstack(lock_eye_imgs))
display(Image(os.path.join(debug_dir,"lock_eye_imgs.png")))
    
cv2.imwrite(os.path.join(debug_dir,"lock_hog_imgs.png"), np.hstack(lock_hog_imgs))
display(Image(os.path.join(debug_dir,"lock_hog_imgs.png")))
```
