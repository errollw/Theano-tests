```python
import cv2
import os
import random
import numpy as np
from IPython.display import Image, display
from skimage import exposure
```

```python
imgs_dir = "C:\Users\Erroll\Documents\datasets\eyespose-20141027"
```

```python
all_imnames = os.listdir(imgs_dir)

ims_matrix_rows = 1
ims_matrix_cols = 10

imrows = []

for i in range(ims_matrix_rows):
    imgs = [cv2.imread(os.path.join(imgs_dir, imname)) for imname in random.sample(all_imnames,ims_matrix_cols)]
    imrows.append(np.hstack(imgs))

display(Image(data=cv2.imencode(".png", np.vstack(imrows))[1].tostring()))
```
