```python
import cv2
import os
import random
import numpy as np
from IPython.display import Image, display, clear_output

from skimage import data, io, filter
```

```python
# from http://www.janeriksolem.net/2009/01/pca-for-images-using-python.html

def pca(X):
    # Principal Component Analysis
    # input: X, each row is a flattened image
    # return: projection matrix (with important dimensions first), variance, and mean

    #get dimensions
    num_data, dim = X.shape

    #center data
    mean_X = X.mean(axis=0)
    for i in range(num_data):
      X[i] -= mean_X

    #only makes sense to return the first num_data
    U,S,V = np.linalg.svd(X)
    V = V[:num_data]

    #return the projection matrix, the variance and the mean
    return V,S,mean_X
```

### Load all images into a matrix

Rows of the matrix are the flattened images.

```python
all_imnames = os.listdir(imgs_dir)
# all_imnames = [imname for imname in os.listdir(dataset_dir_single) if "_g" in imname]

# make a balanced dataset
nolock_fns = [fn for fn in all_imnames if "nolock" in fn]
lock_fns = set(all_imnames) - set(nolock_fns)
num_lock_imgs = len(lock_fns)
nolock_fns_balanced = random.sample(nolock_fns, num_lock_imgs)

all_imnames_balanced = list(set(nolock_fns_balanced).union(lock_fns))

all_ims = [cv2.imread(os.path.join(imgs_dir, imname), cv2.CV_LOAD_IMAGE_GRAYSCALE) for imname in all_imnames_balanced]
# all_ims = [cv2.resize(im, (40,24)) for im in all_ims]
all_ims_flat = [im.flatten() for im in all_ims]

matrix = np.vstack(all_ims_flat)

print matrix.shape
```

### Perform PCA and display the mean image

```python
# perform PCA
V, S, immean = pca(matrix)

mean_img = immean.reshape(24,40)
cv2.normalize(mean_img, mean_img, 0, 255, cv2.NORM_MINMAX)

Image(data=cv2.imencode(".png", mean_img)[1].tostring())
```

### Display eigenvectors of dataset

```python
princom_imgs = []

for eigen_vec in V[:16]:
    eigen_vec_img = eigen_vec.reshape(24,40).astype(np.float32)
    cv2.normalize(eigen_vec_img, eigen_vec_img, 0, 255, cv2.NORM_MINMAX)
    princom_imgs.append(eigen_vec_img)

Image(data=cv2.imencode(".png", np.hstack(princom_imgs))[1].tostring())
```

### Display eye images reconstructed

```python
idxs = random.sample(range(len(all_ims_flat)), 16)

orig_imgs, reconstructed_imgs = [], []

for idx in idxs:
    
    coeffs = [np.dot(all_ims_flat[idx] - immean, v) for v in V[:150]]

    reconstructed_img = immean.copy()
    for (coeff, vec) in zip(coeffs,V[:150]):
        reconstructed_img += coeff * vec
        
    orig_imgs.append(all_ims_flat[idx].reshape(24,40))
    reconstructed_imgs.append(reconstructed_img.reshape(24,40).reshape(24,40))

cv2.imwrite(os.path.join(debug_dir,"orig_imgs.png"), np.hstack(orig_imgs))
display(Image(os.path.join(debug_dir,"orig_imgs.png")))
cv2.imwrite(os.path.join(debug_dir,"reconstructed_imgs.png"), np.hstack(reconstructed_imgs))
display(Image(os.path.join(debug_dir,"reconstructed_imgs.png")))
```

```python
# normalize the eigenvalues
eigenvals = S/S.sum()

# work out how many components needed for decent percentage of variance
for l in range(len(eigenvals)):
    if eigenvals[:l].sum() > 0.6:
        print "60% of variance expressed by top " + str(l) + " components"
        break
```
